```python
import pandas as pd
import sklearn
from flair.models import TextClassifier
from flair.data import Sentence
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import json
from googlesearch  import search
import yfinance as yahooFinance
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import precision_recall_fscore_support

sia = TextClassifier.load('en-sentiment')
sid_obj = SentimentIntensityAnalyzer()
sns.set() 
```

### Load Data

```python
data = pd.read_csv("Data/elon_musk_tweets_ner.csv")
```

```python
data.head()
```

### Extraction

#### Flair

```python
# use Flair to extract sentiment of tweets

def flair_prediction(x):
    sentence = Sentence(x)
    sia.predict(sentence)
    score = sentence.labels[0]
    if "POSITIVE" in str(score):
        return "pos"
    elif "NEGATIVE" in str(score):
        return "neg"
    else:
        return "neu"
    
data["sentiment"] = data.tweet.apply(flair_prediction)
```

```python
# use Flair to extract sentiment score of tweets
def flair_prediction_score(x):
    sentence = Sentence(x)
    sia.predict(sentence)
    return sentence.score
    
data["sentiment_score_flair"] = data.tweet.apply(flair_prediction_score)
```

#### Vador

```python
# use Vador to extract sentiment of tweets

def vador_prediction(x):
    sentiment_dict = sid_obj.polarity_scores(x)
    if sentiment_dict['compound'] > 0 :
        return "pos"
    elif sentiment_dict['compound'] < 0 :
        return "neg"
    else :
        return "neu"
        
data["sentiment_vader"] = data.tweet.apply(vador_prediction)
```

```python
# save data
data.to_csv('Data/ner_tags_hugging_face_sentiment.csv', index=False)
```

#### Kappa score between Flair and Vador

```python
# cohen kappa score between Flair an Vader sentiment analysis
sklearn.metrics.cohen_kappa_score(data.sentiment, data.sentiment_vader, labels=None, weights=None, sample_weight=None)
```

### Extract tickers that have data in Yahoo FInance

```python
flair_ner = pd.read_csv('Data/ner_tags_flair.csv')
```

```python
flair_ner.loc[flair_ner.NER == 'ORG'].WORD.value_counts()
```

```python
# convert comapny name to potential ticker from google first result
def name_convert(self):

    searchval = 'yahoo finance '+self
    link = []
    #limits to the first link
    for url in search(searchval, tld='es', lang='es', stop=1):
        link.append(url)

    link = str(link[0])
    link=link.split("/")
    if link[-1]=='':
        ticker=link[-2]
    else:
        x=link[-1].split('=')
        ticker=x[-1]

    return(ticker)
```

```python
# check from potential tcikers which have data in yahoo and save them in a dictionay
dict_companies_tickers = {}
for company in set(list(flair_ner.loc[flair_ner.NER == 'ORG'].WORD)):
    ticker = name_convert(company)
    temp = yf.download(ticker, 
                      start='2019-01-01', 
                      end='2019-02-12', 
                      progress=False,
    )
    if len(temp) > 0:
        dict_companies_tickers[company] = ticker
```

```python
# save company-ticker dict to json
a_file = open("Data/company_to_ticker.json", "w")

json.dump(dict_companies_tickers, a_file)

a_file.close()
```

### Vizualize sentiment and compare it to stock price

```python
# use moving average window to plot sentiment score over time
def movingaverage(interval, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')


y_av = movingaverage(data.loc[data.date > '2016-01-01'].sentiment_score_flair, 20)
fig = plt.figure(figsize = (30,5))
plt.plot_date(x=data.loc[data.date > '2016-01-01'].date, y=y_av, fmt='-')
#plt.plot_date(x=aapl_df.index, y=aapl_df.Open, fmt='-')

plt.ylabel('Adjusted close')
plt.xlabel('Date')
plt.show()
```

```python
# calculate positive/negative ratio per year
ratios = {}

for x in range(2015, 2022):
    pos = data.loc[(data.date > str(x) + '-01-01') & (data.date < str(x+1) + '-01-01')].sentiment.value_counts()['pos']
    neg = data.loc[(data.date > str(x) + '-01-01') & (data.date < str(x+1) + '-01-01')].sentiment.value_counts()['neg']
    ratio = pos/neg
    ratios[str(x) ] = ratio
```

```python
# plot positive/negative ratio per year
rations

fig = plt.figure(figsize = (15,5))
plt.plot_date(x=ratios.keys(), y=ratios.values(), fmt='-')
#plt.plot_date(x=aapl_df.index, y=aapl_df.Open, fmt='-')

plt.title('Ration Positive/Negative sentiment per year')
plt.ylabel('Positive/Negative')
plt.xlabel('Date')
plt.show()
```
